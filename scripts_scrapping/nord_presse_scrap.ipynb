```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as soup
import re
import pandas as pd
import numpy as np

from datetime import datetime,date,timedelta  

import time
```

```python
start_time = time.time()

day_stop = pd.to_datetime('01-01-2021' , format="%d-%m-%Y").date()
day = pd.to_datetime('01-01-2016' , format="%d-%m-%Y").date()


headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'}
liste_titre = []
liste_date = []
jour_sans_titre = []

while day!= day_stop:
    try:
        day_str = day.strftime('%Y/%m/%d/')
        url = "https://nordpresse.be/"+day_str
        req = Request(url , headers=headers)
        page = soup( urlopen(req).read() , "lxml")
        
        liste_balise = page.find('ul',{'class':'mvp-blog-story-list left relative infinite-content'}).findAll('h2')
        for titre in liste_balise:
            liste_titre.append(titre.getText().encode('utf-8').decode('utf-8').replace(u'\xa0', u' '))
            liste_date.append(day)
        
        
        
        print(day," Temps Exec°: " ,(time.time() - start_time))
        day = day + timedelta(days=1)
        
    except Exception:
        jour_sans_titre.append(day)
        day = day + timedelta(days=1)
```

```python
df_np = pd.DataFrame()
#liste_titre = []
#liste_titre.append(pd.to_datetime(day , format="%d-%m-%Y").date())
df_np['titre'] = liste_titre
df_np['date'] = liste_date
df_np.to_csv(index=False ,path_or_buf='data_scrapping_csv/nordpresse.csv', encoding = 'utf-8')
```
