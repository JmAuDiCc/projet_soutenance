```python
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup as soup
import re
import pandas as pd
import numpy as np

from datetime import datetime,date,timedelta  

import time

import dateparser
```

```python
liste_themes = ['politique','international','societe','ecologie','economie','sciences','culture','religions']
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'}
start_time = time.time()

liste_titre = []
liste_date = []
liste_th = [] #theme de chaques art

day_stop = pd.to_datetime('01-01-2021' , format="%d-%m-%Y").date()

for theme in liste_themes:
    print(theme)
    e = False
    numero_page = 1
    while not e :
        e = False #condition sortie ==> pas de page et date == demain
        url = "https://charliehebdo.fr/themes/"+theme+'/page/'+str(numero_page)+'/'
        req = Request(url , headers=headers)
        
        if numero_page % 5 ==0:
            print(numero_page)
        
        try:
            page = soup( urlopen(req).read() , "lxml")
            ruban =  page.find('ul',{"class":'d-flex flex-row flex-wrap p-0 m-0'})
            titres = ruban.findAll('h3')
            dates = ruban.findAll('span',{'class':'ch_post_date content-block__author'})
            for titre,date in zip (titres,dates):
                day = dateparser.parse(date.getText().encode('utf-8').decode('utf-8').replace(u'\xa0', u' ')).date()
                if day <= day_stop:
                    liste_titre.append(titre.getText().encode('utf-8').decode('utf-8').replace(u'\xa0', u' '))
                    liste_date.append(day)
                    liste_th.append(theme)
                    
            numero_page +=1
            
        except Exception: #pas de page (Erreur 404)
            print(theme," Temps Exec°: " ,(time.time() - start_time))
            e = True
```

```python
df_c = pd.DataFrame()
#liste_titre = []
#liste_titre.append(pd.to_datetime(day , format="%d-%m-%Y").date())
df_c['titre'] = liste_titre
df_c['date'] = liste_date
df_c['theme'] = liste_th
#df_c.to_csv(index=False ,path_or_buf='data_scrapping_csv/charlie_hebdo.csv', encoding = 'utf-8')
```
